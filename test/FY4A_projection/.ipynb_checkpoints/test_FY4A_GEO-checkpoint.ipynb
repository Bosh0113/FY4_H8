```python
from fy4ageo import AGRI_L1
import numpy
import matplotlib.pyplot as plt
```

```python
hdf_path = '/disk2/workspace/20230814/FY4A-_AGRI--_N_DISK_1047E_L1-_GEO-_MULT_NOM_20220624061500_20220624062959_4000M_V0001.HDF'
```

```python
# 可视化原始圆盘影像
import h5py
def display_original_DF(hdf_path):
    f = h5py.File(hdf_path, 'r')
    dset = f['NOMSatelliteZenith'] # DN值
    d_array = numpy.array(dset)
    d_array[d_array>=65534.] = numpy.NaN
    f.close()
    plt.imshow(d_array, interpolation=None)
    plt.colorbar()
```

```python
display_original_DF(hdf_path)
```

```python
def proj_clip_ex(hdf_path, geo_desc, band_name):
    fy4a_file = AGRI_L1(hdf_path, geo_desc)
    geo_dn = fy4a_file.extract(band_name)
    return geo_dn
```

```python
geo_desc = [-59.99, 60, 44, 163.99, 0.04]  # 顺序为南、北、西、东、分辨率，即[lat_s, lat_n, lon_w, lon_e, resolution] 中心104.7°
vza_AHI = proj_clip_ex(hdf_path, geo_desc, 'SatelliteZenith')
s_array = numpy.array(vza_AHI)
```

```python
s_array.shape
```

```python
# 可视化裁切投影区域影像
plt.imshow(s_array, interpolation=None)
plt.colorbar()
```

```python
geo_desc = [-59.99, 60, 44, 163.99, 0.04]  # 顺序为南、北、西、东、分辨率，即[lat_s, lat_n, lon_w, lon_e, resolution] 中心104.7°
vaa_AHI = proj_clip_ex(hdf_path, geo_desc, 'SatelliteAzimuth')
vaa_AHI = numpy.array(vaa_AHI)
# 可视化裁切投影区域影像
plt.imshow(vaa_AHI, interpolation=None)
plt.colorbar()
```

## Save Other Satellite VZA VAA

```python
def save_angle(geo_desc, angle_type, storage_path):
    angle_cliped = proj_clip_ex(hdf_path, geo_desc, angle_type)
    angle_cliped = numpy.array(angle_cliped)
    numpy.save(storage_path, angle_cliped)
    return angle_cliped
```
